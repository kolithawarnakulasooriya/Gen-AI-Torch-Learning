# Finding Similar Movies

```python
# Imports
import pandas as pd
import numpy as np

repo = 'https://raw.githubusercontent.com/kolithawarnakulasooriya/Gen-AI-Torch-Learning/main/asset/ml-100k/' 
```

```python
# Movie data
m_cols = ['movieId','title']
movies = pd.read_csv(repo + 'u.item', sep='|', names=m_cols, usecols=range(len(m_cols)), encoding='latin-1')
movies.head()
```

```python
# Ratings data
r_cols = ['userId', 'movieId','rating']
ratings = pd.read_csv(repo + 'u.data', sep='\t', names=r_cols, usecols=range(len(r_cols)))
ratings.head()
```

```python
# Merge All 
movie_ratings = pd.merge(ratings, movies)
movie_ratings.head()
```

```python
movie_ratings_pivot = pd.pivot_table(movie_ratings, index='userId', columns=['title'], values='rating')

movie_ratings_pivot.head()
```

```python
m1_ratings = movie_ratings_pivot['20,000 Leagues Under the Sea (1954)']
m1_ratings
```

```python
# Find similalirity
"""
    Use corrwith to find similalirity index
    
    |x|y|               |t|              
    |1|4|  Core With    |3|
    |2|5|               |4|
    
"""
similar_movies = movie_ratings_pivot.corrwith(m1_ratings).dropna()
similar_movies.head()
```

```python
similar_movies.sort_values(ascending=False)
```

```python
# calculate average ratings

avg_ratings = movie_ratings.groupby('title').agg({'rating':[np.size, np.mean]})
avg_ratings.head()
```

```python
# remove underratted movies

avg_ratings = avg_ratings[avg_ratings['rating']['size'] >= 100]
avg_ratings.sort_values([('rating','mean')], ascending=False).head()
```

```python
ratings_with_avg_ratings = avg_ratings.join(pd.DataFrame(similar_movies))
ratings_with_avg_ratings.head()
```
